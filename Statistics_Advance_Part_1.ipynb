17. What are point estimates and interval estimates in statistics?
- Point Estimate
- A point estimate is a single value calculated from sample data used to estimate an unknown population parameter.

- It provides a specific numeric value as the best guess of the parameter.

| --------------------- | --------------------------------- |
| Population mean       | Sample mean $\bar{x} = 50$        |
| Population proportion | Sample proportion $\hat{p} = 0.6$ |
| Population variance   | Sample variance $s^2 = 25$        |


- Interval Estimate:

- An interval estimate provides a range of values within which the population parameter is expected to lie.

- It accounts for sampling variability and expresses the uncertainty of the estimate.

- This range is usually expressed as a confidence interval with a specified confidence level (e.g., 95%).

- General form:

18. What is the significance of confidence intervals in statistical analysis?
- A confidence interval is a range of values, calculated from sample data, that is likely to contain the true population parameter (such as a mean or proportion) with a specified level of confidence (e.g., 90%, 95%, or 99%).

- Significance of Confidence Intervals:

 1. Estimate Precision
CIs provide a range instead of a single point estimate, giving a sense of how precise the estimate is. A narrow interval means more precision; a wide interval means less.

 2. Quantify Uncertainty
They quantify the uncertainty inherent in using sample data to estimate population parameters, showing the degree of confidence statisticians have in the estimate.

 3. Decision-Making Tool
CIs help in making informed decisions by showing the range of plausible values for the parameter. For example, in medicine, it helps determine if a treatment effect is significant or likely due to chance.

 4. Inferential Statistics Basis
They allow researchers to generalize results from a sample to the whole population with a known confidence level.

 5. Hypothesis Testing Support
CIs can be used to test hypotheses. If a CI for a mean difference does not include zero, it indicates a statistically significant difference.

- We are 95% confident that the true mean reduction lies between 5 and 15 units.

21. What are the assumptions for applying the Central Limit Theorem?
- The Central Limit Theorem (CLT) states that, regardless of the shape of the population distribution, the sampling distribution of the sample mean approaches a normal distribution as the sample size becomes large.

- But to apply CLT accurately, certain assumptions must be satisfied

- Key Assumptions of the Central Limit Theorem:
1. Independence of Observations

Each sample (or observation) must be independent of the others.

Example: The outcome of one coin flip doesn't affect another.

🔹 Violation Example: Sampling from a population without replacement in a small dataset.

2. Identically Distributed Variables

The data should come from the same population and follow the same probability distribution.

Each variable should have the same mean and variance.

3. Sample Size Should Be Sufficiently Large

A larger sample size (n ≥ 30) generally makes the CLT more reliable.

If the original distribution is heavily skewed, even larger samples may be needed.

4. Finite Variance and Mean

The population must have a finite (i.e., well-defined and not infinite) mean
𝜇
μ and variance
𝜎
2
σ
2
 .

Distributions with infinite variance (like Cauchy distribution) do not follow CLT.

| Assumption               | Description                                   |
| ------------------------ | --------------------------------------------- |
| Independence             | Observations must not affect each other       |
| Identical Distribution   | All data must come from the same distribution |
| Large Sample Size        | n ≥ 30 is a rule of thumb                     |
| Finite Mean and Variance | Population must not have infinite values      |


- Real-World Example:
- Suppose you're measuring the average delivery time for a product. Even if delivery times are not normally distributed, if you:
